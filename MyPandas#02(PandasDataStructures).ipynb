```python
import numpy as np
import pandas as pd
##There are two basic kinds of data structures in Python-Pandas
```

```python
## 1. Series -> A single dimensional data storage medium
ser_1 = pd.Series(np.random.rand(20))
ser_1
```

```python
ser_2 = pd.Series([1,2,3,4,5,6,7,8])
ser_2
```

```python
## 2. DataFrame is a 2D data storage medium
my_df1 = pd.DataFrame(np.random.rand(50,5),index=np.arange(50))
my_df1
```

```python
##to delete a column from a dataframe we use drop function and put the 2nd argument axis=1(this shows column should be deleted)
my_df1.drop(0,axis=1)
my_df1.head()
```

```python
##similarly if we want to delete a row from a dataframe we put axis=0 in drop()
my_df1.drop(1,axis=0) ##this shows that row 1 is deleted from the dataframe
my_df1.head()
```

```python
##if you want to create copy of a dataframe, you can use copy() function
my_df2 = my_df1.copy()
my_df2.head()
##keep one thing in mind that if we directly assign a dataframe to a new dataframe 
##variable it will not create a copy rather points to the same dataframe i.e my_df2 = my_df1
##this will not create a copy of my_df1 in my_df2 insead my_df2 will starts pointing my_df1 
```

```python

my_df2.drop((0,0),axis=1)
```

```python
##if we want to alter a value at a particular location in a dataset, we can do this
my_df2.loc[1,0] = 0.8027384
```
